## Sentence Transformer on HuggingFace
SentenceTransformers is a Python framework for state-of-the-art sentence, text and image embeddings. Install the Sentence Transformers library. one of the embedding models is used in HuggingFaceEmbedding class. 

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "this is test documents"
query_result = embedding.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embedding.embed_documents([text, "This is not a test document."])
doc_result
```
